```python
import pandas as pd

# Provide the correct path to your CSV file
file_path = "98-401-X2021007_eng_CSV/98-401-X2021007_English_CSV_data.csv"


df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display the first few rows of the dataset
df.head()
```

```python
import json

# Load your geojson file
with open('census_Boundaries.geojson', 'r') as f:
    geojson_data = json.load(f)

# Extract the unique DGUID values
dguids = set()

for feature in geojson_data['features']:
    dguids.add(feature['properties']['DGUID'])

# Convert set to list (optional, if you need it as a list)
unique_dguids = list(dguids)


filtered_df = df[df['DGUID'].isin(unique_dguids)]

filtered_df
```

```python
# Convert unique_dguids (from GeoJSON) to a set for better performance
unique_dguids_set = set(unique_dguids)

# Filter the DataFrame to only include rows where DGUID is in unique_dguids
matching_df = df[df['DGUID'].isin(unique_dguids_set)]

# Get the number of unique DGUIDs from unique_dguids that are present in the df
num_unique_dguids_in_df = matching_df['DGUID'].nunique()

# Calculate the percentage of unique DGUIDs present in df
percentage_in_df = (num_unique_dguids_in_df / len(unique_dguids)) * 100

# Output the percentage
print(f"Percentage of unique DGUIDs from GeoJSON that are in df: {percentage_in_df:.2f}%")
```

```python
filtered_df.to_csv('cencusdataunclean.csv', index=False)  # index=False to avoid saving the index column
```

```python

import math

# Define the input file and output prefix
input_file = "cencusdataunclean.csv"
output_prefix = "Raw_census_part_"
num_parts = 6  # Number of parts to split the file into

# Read the entire CSV to determine the number of rows
try:
    df = pd.read_csv(input_file)
    total_rows = len(df)
    rows_per_part = math.ceil(total_rows / num_parts)

    # Split the DataFrame into 4 parts and save each as a CSV
    for i in range(num_parts):
        start_row = i * rows_per_part
        end_row = min((i + 1) * rows_per_part, total_rows)  # Ensure we don't go out of bounds
        chunk = df.iloc[start_row:end_row]
        output_file = f"{output_prefix}{i+1}.csv"
        chunk.to_csv(output_file, index=False)
        print(f"Part {i+1} saved to {output_file}, rows {start_row} to {end_row - 1}")

except Exception as e:
    print(f"An error occurred: {e}")
```
